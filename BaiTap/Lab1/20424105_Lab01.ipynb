```python
import pandas as pd
import numpy as np
import random as rd
```

```python
pd.set_option('display.max_columns', 10)

train_data = pd.read_csv("train.csv", encoding='utf-8')

print(train_data.info())
```

```python
print(train_data.describe())
```

Theo như dữ liệu thì ta thấy thuộc tính Cabin có 204 dòng, Age có 714 dòng, Embarked có 889 dòng. Các thuộc tính này hiện đang bị khuyết. 
Bây giờ chúng ta thực hiện việc xóa các cột không ảnh hướng đến việc phân tích:

```python
train_data.drop(['PassengerId', 'Ticket', 'Cabin'], axis = 1, inplace = True)
```

Xuất thông tin các cột còn lại.

```python
train_data.iloc[0:10]
```

Tiếp theo ta xóa dữ liệu khuyết.

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace = True)

train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace = True)

train_data['Fare'].fillna(train_data['Fare'].median(), inplace = True)
```

```python
train_data.iloc[0:10]
```

```python
train_data['People_Number'] = train_data['SibSp'] + train_data['Parch'] + 1

train_data['Name_c'] = train_data['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]

train_data['Fare_c'] = pd.qcut(train_data['Fare'], 4)

train_data['Age_c'] = pd.cut(train_data['Age'].astype(int), 5)

print(train_data.info())
```

```python
train_data.iloc[0:0,10]
```

Chúng ta sẽ quan tâm đến hai đặc trưng là Sex và Embarked. Đầu tiên ta sẽ biến đổi các thuộc tính này thành biểu diễn dạng số học.

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

name_le = LabelEncoder()
name_labels = name_le.fit_transform(train_data['Name_c'].astype(str))
train_data['Name_Label'] = name_labels

sex_le = LabelEncoder()
sex_labels = sex_le.fit_transform(train_data['Sex'].astype(str))
train_data['Sex_Label'] = sex_labels

fare_le = LabelEncoder()
fare_labels = fare_le.fit_transform(train_data['Fare_c'].astype(str))
train_data['Fare_Label'] = fare_labels

age_le = LabelEncoder()
age_labels = age_le.fit_transform(train_data['Age_c'].astype(str))
train_data['Age_Label'] = age_labels

emb_le = LabelEncoder()
emb_labels = emb_le.fit_transform(train_data['Embarked'].astype(str))
train_data['Embarked_Label'] = emb_labels

poke_df_sub = train_data[['Survived', 'Pclass', 'Name_Label', 
                            'Sex_Label', 'Age_Label', 'SibSp', 
                            'Parch', 'Fare_Label', 'Embarked_Label', 
                            'People_Number']]
print(poke_df_sub.info())
```

Sau khi biểu diễn xong ta các cột dưới đây mô tả dưới dạng số học. Tiếp theo chúng ta sẽ sử dụng One-hot Encoding cho các đặc trưng này.

```python
#Name
name_ohe = OneHotEncoder()
name_feature_arr = name_ohe.fit_transform(
                                poke_df_sub[['Name_Label']]).toarray()
name_feature_labels = ['Name_' + str(cls_label) 
                           for cls_label in name_le.classes_]
name_features = pd.DataFrame(name_feature_arr, 
                            columns = name_feature_labels)
#Sex
sex_ohe = OneHotEncoder()
sex_feature_arr = sex_ohe.fit_transform(
                                poke_df_sub[['Sex_Label']]).toarray()
sex_feature_labels = ['Sex_' + str(cls_label) 
                           for cls_label in sex_le.classes_]
sex_features = pd.DataFrame(sex_feature_arr, 
                            columns = sex_feature_labels)
#Fare
fare_ohe = OneHotEncoder()
fare_feature_arr = fare_ohe.fit_transform(
                                poke_df_sub[['Fare_Label']]).toarray()
fare_feature_labels = ['Fare_' + str(cls_label) 
                           for cls_label in fare_le.classes_]
fare_features = pd.DataFrame(fare_feature_arr, 
                            columns = fare_feature_labels)
#Age
age_ohe = OneHotEncoder()
age_feature_arr = age_ohe.fit_transform(
                                poke_df_sub[['Age_Label']]).toarray()
age_feature_labels = ['Age_' + str(cls_label) 
                           for cls_label in age_le.classes_]
age_features = pd.DataFrame(age_feature_arr, 
                            columns = age_feature_labels)

#Embarked
emb_ohe = OneHotEncoder()
emb_feature_arr = emb_ohe.fit_transform(
                                poke_df_sub[['Embarked_Label']]).toarray()
emb_feature_labels = ['Embarked_' + str(cls_label) 
                           for cls_label in emb_le.classes_]
emb_features = pd.DataFrame(emb_feature_arr, 
                            columns = emb_feature_labels)
```

Bây giờ thì ta gọp các đặc trưng được tạo ra ở trên lại.

```python
poke_df_ohe = pd.concat([poke_df_sub, name_features, 
                         sex_features, fare_features, 
                         age_features, emb_features], axis=1)
# columns = sum([['Survived', 'Pclass'], 
#                name_feature_labels,
#                sex_feature_labels, 
#                fare_feature_labels,
#                age_feature_labels,
#                ['SibSp', 'Parch'],
#                emb_feature_labels], [])
```

```python
poke_df_ohe.drop(['Name_Label', 'People_Number', 'Sex_Label', 'Age_Label'
                  , 'Embarked_Label', 
                 'Fare_Label'], axis = 1, inplace = True)
print(poke_df_ohe.info())
```

Cho 5 xúc xắc, lắc 1000 lần. Tính tỉ lệ ra mặt 1 cho cả 5 xúc xắc.

```python
def laMatMotDiem():
    return rd.randint(1, 6) == 1
```

```python
def tungXucXac():
    for i in range(5):
        if laMatMotDiem():
            continue
        return False
    return True
```

```python
def tungNLan(n):
    dem = 0
    for i in range(n):
        if tungXucXac():
            dem += 1
    return dem/n
```

```python
print(f'Tỉ lệ tung được 5 xúc xắc mặt 1 là: {tungNLan(1000)}%')
```
