#### Bài 1: Cho 5 con xúc xắc, lắc 1000 lần. Tính tỉ lệ ra mặt 1 cho cả 5 xúc xắc.

```python
import pandas as pd
import numpy as np
import random as rd
```

```python
def LaMatMotDiem():
    return rd.randint(1, 6) == 1
```

```python
def TungXucXac():
    for i in range(5):
        if LaMatMotDiem():
            continue
        return False
    return True
```

```python
def TungNLan(n):
    dem = 0
    for i in range(n):
        if TungXucXac():
            dem += 1
    return dem/n
```

```python
print(f'Tỉ lệ tung 5 con xúc xắc mặt 1 là: {TungNLan(1000)}%')
```

Điền dữ liệu khuyết

```python
data['Age'].fillna(data['Age'].median(), inplace = True)

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace = True)

data['Fare'].fillna(data['Fare'].median(), inplace = True)
```

Mã hóa dán nhãn (LabelEncoder) với 2 đặc trưng Sex, Embarked kiểu dữ liệu là object. 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Sex
sex_le = LabelEncoder() 
data['Sex']= sex_le.fit_transform(data['Sex'].astype(str)) 

# Embarked
emb_le = LabelEncoder()
data['Embarked'] = emb_le.fit_transform(data['Embarked'].astype(str))
```
