## Đề: Xấp xỉ số pi bằng phương pháp Monte Carlo
### Ta có:
* #### Số kim trong hình tròn: Nc
* #### Số kim trong hình vuông: Ns
#### Phương pháp Monte Carlo
<h3><center> $\pi = 4\frac{N_c}{N_s}$</center></h3>

- Để biết được kim đó có nằm trong hình tròn hay không, giả sử bán kính = 1, thì ta tính xem khoảng cách từ A đến B (kim) vừa mới rải có <= 1 hay không. Ở đây ta sử dụng định lý Pytago để kiểm tra.
$$C = \sqrt{A^2 + B^2}$$

```python
import math
import random as rd
import matplotlib.pyplot as plt
```

```python
def hypotenuse(a, b):
  return math.sqrt(math.pow(a,2) + math.pow(b,2))
```

```python
def pi(n):
    Ns = 0 
    Nc = 0

    while Ns < n:
        x = rd.random()
        y = rd.random()
        if(hypotenuse(x, y) <= 1):
            Nc += 1
            plt.plot(x , y , label= "stars", color= "blue", marker= "*")
        Ns += 1;
        plt.plot(x , y , label= "point", color= "red",  marker= ".")
        
    pi = (Nc * 4.0) / Ns
   
    return pi
```

```python
n = float(input("Nhập số điểm: "))
pi = pi(n)
print(f'Số pi: {pi}')

# plt.show()
```
